```python
import pandas as pd

def get_table_data():
    conn = connect_to_db()
    query = 'SELECT "id", "title", "GTIN" FROM "ManufSKUs" WHERE "GTIN" IS NOT NULL'
    df_db = pd.read_sql(query, conn)
    conn.close()
    return df_db

# Пример использования
df_db = get_table_data()
df_db['GTIN']
```

```python
# Функция для поиска совпадений GTIN и добавления информации
def add_info(row, df_db):
    g_list = row['GTIN'].split()
    ids, titles, gtins = [], [], []
    for gtin in g_list:
        matched_rows = df_db[df_db['GTIN'].apply(lambda x: any(gtin in item for item in x))]
        if not matched_rows.empty:
            ids.extend(matched_rows['id'].tolist())
            titles.extend(matched_rows['title'].tolist())
            gtins.extend([' '.join(x) for x in matched_rows['GTIN'].tolist()])
    if ids:
        row['DB_ids'] = ' '.join(map(str, ids))
        row['DB_titles'] = ' '.join(titles)
        row['DB_GTINs'] = ' '.join(gtins)
    else:
        row['DB_ids'] = '-'
        row['DB_titles'] = '-'
        row['DB_GTINs'] = '-'
    return row

# Добавление колонок в DataFrame
df_GTIN['DB_ids'] = None
df_GTIN['DB_titles'] = None
df_GTIN['DB_GTINs'] = None

# Применение функции для добавления информации
df_GTIN = df_GTIN.apply(lambda row: add_info(row, df_db), axis=1)

# Вывод первых нескольких строк данных для проверки
print("First few rows of the updated df_GTIN:")
print(df_GTIN.head())

# Проверка на наличие совпадений
if df_GTIN['DB_ids'].str.contains('-').all():
    print("No matches found")
else:
    print("Matches found and added to df_GTIN")
```
