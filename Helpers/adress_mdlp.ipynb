```python
import pandas as pd
from fuzzywuzzy import fuzz
from collections import Counter

# Функция для чтения данных из Excel файла с сохранением ведущих нулей в branch_id
def read_data(file_path):
    return pd.read_excel(file_path, dtype={'branch_id': str})  # Сохраняем ведущие нули для branch_id

# Функция для сравнения адресов
def compare_addresses(address_desc, physical_address):
    match_score_physical = fuzz.token_sort_ratio(address_desc, physical_address)
    return match_score_physical

# Основная функция для обработки файла и сравнения адресов
def process_addresses(file_path, threshold=80):
    # Чтение данных из файла Excel
    df = read_data(file_path)

    results = []

    for index, row in df.iterrows():
        address_desc = row['address_desc']
        physical_address = row['physicalAddress']
        
        match_score_physical = compare_addresses(address_desc, physical_address)

        # Сохранение результатов для дальнейшего анализа
        results.append({
            "branch_id": row['branch_id'],
            "id": row['id'],
            "address_desc": address_desc,
            "physicalAddress": physical_address,
            "match_score_physical": match_score_physical
        })

    # Преобразуем результаты в DataFrame для удобного анализа
    results_df = pd.DataFrame(results)

    # Группируем по branch_id и анализируем совпадения
    final_decision = []
    for branch_id, group in results_df.groupby('branch_id'):
        # Собираем все совпадения для physicalAddress
        physical_matches = group['match_score_physical'].tolist()

        # Считаем, какой результат совпадения встречается чаще всего
        most_common_physical = Counter(physical_matches).most_common(1)[0]

        # Добавляем решение для каждого branch_id, независимо от порога
        final_decision.append({
            "branch_id": branch_id,
            "common_match_physical": most_common_physical,
            "decision": "Нужно вставить новый адрес" if most_common_physical[0] < threshold else "Совпадение выше порога, адрес корректен"
        })

    # Преобразуем результаты в DataFrame для удобного вывода
    final_df = pd.DataFrame(final_decision)

    return final_df

# Путь к файлу Excel
file_path = '/mnt/d/Doc_To_Date/20240923/Result_171.xlsx'

# Обрабатываем файл и получаем результаты анализа
results_df = process_addresses(file_path)


```
