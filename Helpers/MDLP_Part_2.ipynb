```python
import pandas as pd
import psycopg2

# Подключаемся к базе данных
conn = psycopg2.connect(
    dbname='pharmhub',
    user='ivan.anisimov',
    password='oA7W5uz2GzEd',
    host='10.0.0.171'
)

# Функция для получения следующего externalId для конкретного client_id
def get_next_external_id(conn, client_id):
    try:
        with conn.cursor() as cursor:
            cursor.execute("""
                SELECT max("externalId")::INT + 1 AS next_id
                FROM "ManufSKUs"
                WHERE client = %s;
            """, (client_id,))
            result = cursor.fetchone()
            if result and result[0]:
                return result[0]
            else:
                return None
    except Exception as e:
        print(f"Ошибка при получении externalId для client_id {client_id}: {e}")
        conn.rollback()  # Откат транзакции при ошибке
        return "ОШИБКА"

# Добавляем новый столбец externalId к DataFrame
df_grouped_nan['externalId'] = None

# Словарь для хранения следующего externalId для каждого производителя
next_external_ids = {}

# Начальное значение для тех, у кого произошла ошибка
error_external_id = 200001

# Обрабатываем каждую строку DataFrame
for i in range(len(df_grouped_nan)):
    client_id = df_grouped_nan.loc[i, 'client_id']
    
    # Если externalId для этого client_id еще не был получен, делаем запрос
    if client_id not in next_external_ids:
        external_id = get_next_external_id(conn, client_id)
        if external_id == "ОШИБКА" or external_id is None:
            # Для ошибок начинаем присваивать externalId с 200001
            next_external_ids[client_id] = error_external_id
            error_external_id += 1
        else:
            next_external_ids[client_id] = external_id
    
    # Присваиваем externalId и инкрементируем для следующих строк
    df_grouped_nan.loc[i, 'externalId'] = next_external_ids[client_id]
    next_external_ids[client_id] += 1

# Закрываем соединение с базой данных
conn.close()

# Выводим результат
print(df_grouped_nan.head())
```

```python
import psycopg2
import pandas as pd

# Подключаемся к базе данных
conn = psycopg2.connect(
    dbname='pharmhub',
    user='ivan.anisimov',
    password='oA7W5uz2GzEd',
    host='10.0.0.171'
)

# Функция для проверки существования производителя в базе
def check_producer_in_db(producer_name):
    try:
        with conn.cursor() as cursor:
            cursor.execute("""
                SELECT C.title, C.id
                FROM "Clients" C
                WHERE C.title ILIKE %s;
            """, (f"%{producer_name}%",))
            results = cursor.fetchall()
            return results if results else None
    except Exception as e:
        print(f"Ошибка при проверке производителя {producer_name}: {e}")
        conn.rollback()
        return None

# Функция для проверки наличия активных SKU у клиента
def has_active_skus(client_id):
    try:
        with conn.cursor() as cursor:
            cursor.execute("""
                SELECT MSU.id
                FROM "ManufSKUs" MSU
                WHERE MSU.client = %s
                AND MSU.status = 'active'
                LIMIT 1;
            """, (client_id,))
            result = cursor.fetchone()
            return result is not None  # True, если есть активные SKU
    except Exception as e:
        print(f"Ошибка при проверке SKU для client_id {client_id}: {e}")
        conn.rollback()
        return False

# Обрабатываем каждого производителя из df_external_2xx и обновляем client_id
def update_producer_ids(df):
    for index, row in df.iterrows():
        producer_name = row['Unnamed: 9']  # Используем столбец с именем производителя
        found_producers = check_producer_in_db(producer_name)  # Ищем производителей в базе
        if found_producers:
            for producer_name_db, producer_id_db in found_producers:
                if has_active_skus(producer_id_db):
                    print(f"Найдено активное SKU для {producer_name_db} (ID: {producer_id_db})")
                    df.at[index, 'client_id'] = producer_id_db  # Обновляем client_id
                    break  # Останавливаем поиск, как только нашли нужного производителя с активными SKU
        else:
            print(f"Производитель {producer_name} не найден в базе данных.")

# Применяем функцию для обновления client_id в df_external_2xx
update_producer_ids(df_external_2xx)

# Закрываем соединение
conn.close()

# Выводим обновленные данные для проверки
print(df_external_2xx[['Unnamed: 9', 'client_id', 'externalId']].head())
```

```python
import psycopg2

# Подключаемся к базе данных
conn = psycopg2.connect(
    dbname='pharmhub',
    user='ivan.anisimov',
    password='oA7W5uz2GzEd',
    host='10.0.0.171'
)

# Функция для получения следующего externalId для конкретного client_id
def get_next_external_id(client_id):
    try:
        with conn.cursor() as cursor:
            cursor.execute("""
                SELECT max("externalId")::INT + 1 AS next_id
                FROM "ManufSKUs"
                WHERE client = %s;
            """, (client_id,))
            result = cursor.fetchone()
            return result[0] if result and result[0] else None
    except Exception as e:
        print(f"Ошибка при получении externalId для client_id {client_id}: {e}")
        conn.rollback()
        return None

# Словарь для хранения последнего externalId для каждого производителя
external_id_tracker = {}

# Функция для обновления данных в DataFrame
def update_producer_data(df, cleaned_producer_name, correct_client_id):
    # Если мы уже обновляли externalId для этого производителя, то продолжаем его инкрементировать
    if correct_client_id in external_id_tracker:
        new_external_id = external_id_tracker[correct_client_id] + 1
    else:
        new_external_id = get_next_external_id(correct_client_id)  # Получаем новый externalId из базы
        if new_external_id is None:
            print(f"Не удалось получить externalId для {cleaned_producer_name}")
            return
    
    # Обновляем значения в DataFrame
    df.loc[df['cleaned_producer'] == cleaned_producer_name, 'Unnamed: 9'] = cleaned_producer_name  # Обновляем производителя
    df.loc[df['cleaned_producer'] == cleaned_producer_name, 'client_id'] = correct_client_id  # Обновляем client_id
    df.loc[df['cleaned_producer'] == cleaned_producer_name, 'externalId'] = new_external_id  # Обновляем externalId

    # Обновляем последний использованный externalId в трекере
    external_id_tracker[correct_client_id] = new_external_id

    print(f"Обновлены данные для производителя {cleaned_producer_name} с client_id {correct_client_id} и externalId {new_external_id}")

# Основная логика для обработки всех производителей
for index, row in df_external_2xx.iterrows():
    cleaned_producer_name = row['cleaned_producer']  # Используем очищенного производителя
    found_producers = check_producer_in_db(cleaned_producer_name)  # Ищем производителей в базе данных
    if found_producers:
        for producer_name_db, producer_id_db in found_producers:
            if has_active_skus(producer_id_db):
                update_producer_data(df_external_2xx, cleaned_producer_name, producer_id_db)
                break  # Останавливаемся, как только нашли правильного производителя с активными SKU

# Закрываем соединение
conn.close()

# Выводим обновленные данные для проверки
print(df_external_2xx[['Unnamed: 9', 'client_id', 'externalId']].head())
```

```python
import psycopg2
import pandas as pd

# Подключаемся к базе данных
conn = psycopg2.connect(
    dbname='pharmhub',
    user='ivan.anisimov',
    password='oA7W5uz2GzEd',
    host='10.0.0.171'
)

# Выполняем запрос для выгрузки данных о производителях
query = """
SELECT C.title, C.id, COUNT(ms.id) AS sku_count
FROM "Clients" C
LEFT JOIN "ManufSKUs" ms ON ms.client = C.id
WHERE C.type = 'manufacturer'
GROUP BY C.id, C.title
ORDER BY C.title;
"""

# Загружаем данные из базы в DataFrame
df_suppliers = pd.read_sql(query, conn)

# Закрываем соединение с базой данных
conn.close()

# Приводим названия к нижнему регистру для сравнения
df_suppliers['title_cleaned'] = df_suppliers['title'].str.lower()

# Группируем по частичному совпадению (например, ключевые слова из названия)
df_suppliers['title_base'] = df_suppliers['title_cleaned'].apply(lambda x: ' '.join(x.split()[:2]))

# Группируем и считаем общие SKU
df_grouped = df_suppliers.groupby('title_base').agg(
    total_skus=pd.NamedAgg(column='sku_count', aggfunc='sum'),
    entries=pd.NamedAgg(column='id', aggfunc='count'),
    client_ids=pd.NamedAgg(column='id', aggfunc=lambda x: ', '.join(x)),
    titles=pd.NamedAgg(column='title', aggfunc=lambda x: ', '.join(x))
).reset_index()

# Отмечаем, кто "пустышка" (если total_skus = 0)
df_grouped['is_empty'] = df_grouped['total_skus'] == 0

# Сохраняем результат в файл для дальнейшего анализа
df_grouped.to_csv('/mnt/d/Doc_To_Date/grouped_suppliers.csv', index=False)

print("Анализ завершен и сохранен в файл grouped_suppliers.csv")
```

```python
import pandas as pd

# Предположим, у нас есть список производителей, загружаем его в DataFrame
# Пример загрузки из файла CSV
df_suppliers = pd.read_csv('/mnt/d/Doc_To_Date/20240927/grouped_suppliers.csv', names=['title', 'sku_count', 'entries', 'client_id', 'full_name', 'is_empty'])

# Приводим названия к нижнему регистру для сравнения
df_suppliers['title_cleaned'] = df_suppliers['title'].str.lower()

# Группируем по частичному совпадению (например, первые два слова в названии)
df_suppliers['title_base'] = df_suppliers['title_cleaned'].apply(lambda x: ' '.join(x.split()[:2]))

# Сортируем по базовому названию
df_sorted = df_suppliers.sort_values(by='title_base')

# Добавляем блоки разделения между разными группами
separator = '____'

# Сохраняем результат в список для добавления разделителей
grouped_list = []
previous_group = None

for index, row in df_sorted.iterrows():
    current_group = row['title_base']
    if current_group != previous_group:
        if previous_group is not None:
            grouped_list.append([separator] * len(row))  # Добавляем разделитель между группами
        previous_group = current_group
    grouped_list.append(row.tolist())  # Добавляем строки в список

# Преобразуем список обратно в DataFrame
df_grouped = pd.DataFrame(grouped_list, columns=df_suppliers.columns)

# Сохраняем результат в файл
df_grouped.to_csv('/mnt/d/Doc_To_Date/grouped_suppliers_with_blocks.csv', index=False)

print("Группировка завершена и сохранена в файл grouped_suppliers_with_blocks.csv")
```

```python
import pandas as pd

# Загрузим CSV-файл
file_path = '/mnt/d/Doc_To_Date/20241003/add_2024_10_03_3.csv'
df = pd.read_csv(file_path, delimiter='|', dtype={'gtin': str})

# Проверим структуру файла
print(df.head())

# Найдем все дублирующиеся externalId внутри одного и того же manuf (client_id)
duplicates = df[df.duplicated(subset=['manuf', 'externalId'], keep=False)]

# Функция для создания уникальных externalId
def make_unique_id(row, duplicate_counts):
    # Получаем текущее количество повторений externalId для данного manuf
    count = duplicate_counts.get((row['manuf'], row['externalId']), 0)
    
    # Увеличиваем счетчик для данного externalId
    duplicate_counts[(row['manuf'], row['externalId'])] = count + 1
    
    # Если это дубликат (count > 0), прибавляем число к externalId
    if count > 0:
        return f"{row['externalId']}_{count + 1}"
    
    # Если это первая встреча externalId, возвращаем его без изменений
    return row['externalId']

# Словарь для хранения количества повторений
duplicate_counts = {}

# Применяем функцию для создания уникальных externalId только для дубликатов
df['externalId'] = df.apply(lambda row: make_unique_id(row, duplicate_counts) if (row['manuf'], row['externalId']) in duplicates[['manuf', 'externalId']].values else row['externalId'], axis=1)

# Проверим, что больше нет дубликатов
if df.duplicated(subset=['manuf', 'externalId']).sum() == 0:
    print("Все дублирующиеся externalId были исправлены.")
else:
    print("Некоторые дубликаты все еще остались.")
```

```python
import psycopg2
from tqdm import tqdm

# Функция для разбивки списка на части
def chunk_list(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Подключаемся к базе данных
conn = psycopg2.connect(
    dbname='pharmhub',
    user='ivan.anisimov',
    password='oA7W5uz2GzEd',
    host='10.0.0.171'
)

cur = conn.cursor()

gtin_list = df_excel['GTIN'].dropna().tolist()

chunks = list(chunk_list(gtin_list, 1000))

found_gtins = []

# Добавляем tqdm для отображения прогресса по обработке каждой части
for chunk in tqdm(chunks, desc="Processing GTIN chunks", unit="chunk"):
    gtin_tuple = tuple(chunk)  # Преобразуем список в кортеж
    
    query = """
    SELECT m."GTIN"
    FROM "ManufSKUs" m
    WHERE EXISTS (
        SELECT 1
        FROM unnest(string_to_array(m."GTIN", ' ')) AS gtin_value
        WHERE gtin_value = ANY(ARRAY[%s])
    );
    """
    
    # Выполняем запрос с текущим "куском" данных
    cur.execute(query, (gtin_tuple,))
    
    # Добавляем найденные значения в общий список
    results = cur.fetchall()
    found_gtins.extend([row[0] for row in results])

# Закрываем соединение с базой данных
cur.close()
conn.close()

# Проверим какие GTIN из Excel отсутствуют в базе данных
missing_gtins = set(gtin_list) - set(found_gtins)

# Выводим отсутствующие GTIN
print("Отсутствующие GTIN:")
print(missing_gtins)
```
