```python
# Extract the first word and remove trailing commas
df_filtered['Лекарственная форма (стандартизованное значение)'] = df_filtered['Лекарственная форма (стандартизованное значение)'].apply(
    lambda x: x.split()[0].replace(',', '')
)

# Display the unique values to verify the result
lek_form_unique_values = df_filtered['Лекарственная форма (стандартизованное значение)'].unique()
print("\nSimplified Unique Release Forms:")
for value in sorted(lek_form_unique_values):
    print(value)
```

```python
# Улучшенная функция для обработки значений в столбце 'Масса/объем в первичной упаковке'
def format_mass_volume(value):
    # Проверяем, является ли значение числом и не является NaN
    if pd.notna(value) and isinstance(value, (float, int)):
        # Если значение является целым (например, 10.0), преобразуем его в целое (10)
        if float(value).is_integer():
            return str(int(value))
        # Если значение имеет дробную часть, округляем до одной значащей цифры и возвращаем как строку (например, 8.4)
        else:
            return f"{value:.1f}".rstrip('0').rstrip('.')
    else:
        # Если значение не является числом, возвращаем его без изменений
        return value

# Применяем функцию к колонке для форматирования значений
df_filtered['Масса/объем в первичной упаковке'] = df_filtered['Масса/объем в первичной упаковке'].apply(format_mass_volume)

# Проверяем уникальные значения, чтобы убедиться, что все прошло успешно
print(df_filtered['Масса/объем в первичной упаковке'].unique())
```

```python
# Define the name of the new column
new_column_name = 'ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА'

# Create the new column with the concatenation logic, adding the additional condition for dosage
df_filtered[new_column_name] = df_filtered.apply(
    lambda row: (
        f"{row['Торговое наименование']} {row['Лекарственная форма (стандартизованное значение)']} {row['Дозировка (нормализованное значение)']}"
        if '~' not in str(row['Дозировка (нормализованное значение)']) and str(row['Дозировка (нормализованное значение)']).strip().upper() != 'НЕ УКАЗАНО'
        else ''
    ),
    axis=1
)

# Display the first few rows to verify the result
print(df_filtered[[new_column_name]].head())
```

```python
# Определим соответствия для замены в словаре
replacements = {
    'СМ[3*];^МЛ': 'МЛ',
    'М[3*]': '-',
    'Л;^ДМ[3*]': 'МЛ',
    'УСЛ. ЕД': 'МЛ',
    'МЛН. МЕ': 'МЕ'
}

# Заменяем значения в колонке 'Название единицы ОКЕИ для потребительской единицы лекарственного препарата'
df_filtered['Название единицы ОКЕИ для потребительской единицы лекарственного препарата'] = df_filtered['Название единицы ОКЕИ для потребительской единицы лекарственного препарата'].replace(replacements)

# Проверим результат, выведем уникальные значения
print(df_filtered['Название единицы ОКЕИ для потребительской единицы лекарственного препарата'].unique())
```

```python
# Название новой колонки для соединения данных
new_column_name = 'ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА'

# Создаем новый столбец с использованием объединения значений из нескольких колонок
def create_full_name(row):
    # Основные части названия
    parts = [
        row['Торговое наименование'], 
        row['Лекарственная форма (стандартизованное значение)'],
        row['Дозировка (нормализованное значение)'],
        row['Масса/объем в первичной упаковке'],
        row['Название единицы ОКЕИ для потребительской единицы лекарственного препарата']
    ]
    
    # Проверяем, нужно ли добавить 'Количество первичной упаковки в потребительской упаковке'
    quantity = row['Количество первичной упаковки в потребительской упаковке']
    if pd.notna(quantity) and quantity != 1:
        parts.append(f"№{quantity}")

    # Создаем итоговую строку, если дозировка не содержит "~" и не равна "НЕ УКАЗАНО"
    if '~' not in str(row['Дозировка (нормализованное значение)']) and str(row['Дозировка (нормализованное значение)']).strip().upper() != 'НЕ УКАЗАНО':
        return ' '.join(map(str, parts)).strip()
    else:
        # Если дозировка содержит "~" или равна "НЕ УКАЗАНО", не добавляем дозировку и добавляем остальные части
        parts_without_dosage = [
            row['Торговое наименование'],
            row['Лекарственная форма (стандартизованное значение)'],
            row['Масса/объем в первичной упаковке'],
            row['Название единицы ОКЕИ для потребительской единицы лекарственного препарата']
        ]
        if pd.notna(quantity) and quantity != 1:
            parts_without_dosage.append(f"№{quantity}")
        return ' '.join(map(str, parts_without_dosage)).strip()

# Применяем функцию к каждой строке DataFrame и создаем новый столбец
df_filtered[new_column_name] = df_filtered.apply(create_full_name, axis=1)

# Проверяем результат
print(df_filtered[[new_column_name]].head(10))
```

```python
# Функция для обновления строк, у которых ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА пусто
def update_empty_full_name(row):
    # Если значение уже заполнено, возвращаем его
    if row['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА']:
        return row['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА']
    
    # Проверяем, если Дозировка (нормализованное значение) содержит "~" или равно "НЕ УКАЗАНО"
    if '~' in str(row['Дозировка (нормализованное значение)']) or str(row['Дозировка (нормализованное значение)']).strip().upper() == 'НЕ УКАЗАНО':
        
        # Строим наименование без дозировки, но с проверкой на пустые и неподходящие значения в единицах ОКЕИ
        parts = [
            row['Торговое наименование'], 
            row['Лекарственная форма (стандартизованное значение)'], 
            row['Масса/объем в первичной упаковке']
        ]
        
        # Добавляем единицу ОКЕИ, только если она не равна "-" и не выглядит неподходящей (например, 'Г' для данного контекста)
        okei_value = row['Название единицы ОКЕИ для потребительской единицы лекарственного препарата']
        if okei_value not in ['-', 'Г', 'ЕД'] and pd.notna(okei_value):
            parts.append(okei_value)

        # Соединяем части с пробелами между ними
        return ' '.join(map(str, parts)).strip()

    # В других случаях возвращаем пустую строку
    return ''

# Применяем функцию к каждой строке для обновления
df_filtered['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА'] = df_filtered.apply(update_empty_full_name, axis=1)

# Проверяем результат
print(df_filtered[['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА']].head(10))
```

```python
# Преобразуем значения в столбце 'ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА' в верхний регистр
df_filtered['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА'] = df_filtered['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА'].str.upper()

# Проверим результат
print(df_filtered[['ПОЛНОЕ НАИМЕНОВАНИЕ ПРЕПАРАТА']].head(10))
```
