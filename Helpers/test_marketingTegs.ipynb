```python
import pandas as pd

# Предположим, что данные уже загружены и выглядят примерно так:
# df = pd.read_excel('/mnt/d/Doc_To_Date/20241007/Result_57.xlsx')

# Извлечем уникальные товары для TRAIN, ограничив до первых 9
unique_items = df['ms.title'].unique()[581:620]
TRAIN = unique_items.tolist()

# Готовим TEST, извлекая по 3 соответствующих вариации для каждого товара
TEST = []
for item in unique_items:
    # Получаем три записи из numbered_checks.title для каждого уникального товара
    variations = df[df['ms.title'] == item]['numbered_checks.title'].head(3).tolist()
    TEST.extend(variations)

# Создание expected_res
expected_res = []
test_id_start = 2001
train_id_start = 1001

# Проходим по каждому элементу TRAIN, создавая 3 записи для каждого из них в expected_res
for i in range(len(TRAIN)):
    train_id = train_id_start + i  # ID для TRAIN
    for j in range(3):
        # Создаем запись expected_res
        expected_res.append({
            'id': test_id_start,  # ID для TEST
            'y': train_id,        # Сопоставление с TRAIN
            'code': 3,
            'proba': False
        })
        test_id_start += 1  # Переходим к следующему ID для TEST

# Проверка результатов
print(f"TRAIN (первые 9 товаров):\n{TRAIN}")
print(f"\nTEST (первые 27 вариаций):\n{TEST}")
print(f"\nexpected_res (первые 27 записей):\n{expected_res}")
```

```python
# Форматируем TRAIN
formatted_train = "TRAIN = [\n"
for item in TRAIN:
    formatted_train += f"    \"{item}\",\n"
formatted_train += "]\n"

# Форматируем TEST
formatted_test = "TEST = [\n"
for item in TEST:
    formatted_test += f"    \"{item}\",\n"
formatted_test += "]\n"

# Форматируем expected_res
formatted_expected_res = "expected_res = [\n"
for item in expected_res:
    formatted_expected_res += f"    {{'id': {item['id']}, 'y': {item['y']}, 'code': {item['code']}, 'proba': {item['proba']}}},\n"
formatted_expected_res += "]\n"

# Выводим в нужном формате
print(formatted_train)
print(formatted_test)
print(formatted_expected_res)
```
