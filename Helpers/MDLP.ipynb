```python
# Функция для выбора между колонками и создания новой строки
def create_combined_string(row):
    product_name = row['Наименование товара на этикетке']
    mass_volume = row['Масса/объем в первичной упаковке']
    dosage = row['Дозировка (стандартизованное значение)']
    manufacturer = row['Производитель готовой ЛФ']
    
    # Проверим, присутствует ли число из дозировки или массы/объема в названии товара
    if any(d in product_name for d in [dosage, mass_volume]):
        # Если дозировка уже есть в названии, добавляем только массу/объем
        if dosage in product_name:
            combined = f"{product_name} {mass_volume} {manufacturer}"
        # Если масса/объем уже есть в названии, добавляем только дозировку
        else:
            combined = f"{product_name} {dosage} {manufacturer}"
    else:
        # Если ни дозировка, ни масса/объем не присутствуют, добавляем оба
        combined = f"{product_name} {mass_volume} {manufacturer}"
    
    return combined
```

```python

import pandas as pd
from fuzzywuzzy import process
from tqdm import tqdm
import re
from unidecode import unidecode

# Список лишних слов
stop_words = set(['ооо', 'зао', 'пвт', 'лц', 'фарм', 'фармацевтика', 'фармацевтикалс', 'лаборатория', 'инк', 'лтд', 'ко', 'с', 'ао', 'ко', 'оао', 'инк',
                  'нпо','мнпк', 'аг', 'аве', 'спа', 'кг', 'гмбх', 'зоо', 'са', 'се', 'нв', 'сас', 'гуз', 'со', 'спк', 'лэбораториз', 'са', 'сау',  'ау',
                  'сас', 'спа', 'срл',
                  ])

# Функция для очистки текста с учетом лишних слов
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)  # Замена множества пробелов на один пробел
    text = re.sub(r'[^а-яa-z0-9\s]', '', text)  # Удаление специальных символов, кроме русских и английских букв, цифр и пробелов
    tokens = re.findall(r'\b\w+\b', text)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(filtered_tokens).strip()
```

```python
import pandas as pd
from tqdm import tqdm
from fuzzywuzzy import process

# Загрузка файлов
unique_manufacturers_df_2 = pd.read_csv('/mnt/d/Doc_To_Date/20240702/unique_matched_manufacturers_v_3.csv')
clients_df = pd.read_csv('/mnt/d/Doc_To_Date/20240703/pharmhub_public_Clients.csv')

# Применение функции очистки к колонкам
unique_manufacturers_df_2['Обработанный тайтл'] = unique_manufacturers_df_2['Производитель готовой ЛФ'].apply(clean_text)
clients_df['title (очищенный)'] = clients_df['title'].apply(clean_text)

# Просмотр первых нескольких строк для проверки
print(unique_manufacturers_df_2.head())
print(clients_df.head())
```

```python
import re

# Список лишних слов
stop_words = set([
    'лтд', 'с', 'а', 'ооо', 'фарма', 'гмбх', 'ао', 'фармасьютикал', 
    'ко', 'п', 'л', 'и', 'зао', 'р', 'инк', 'оао', 'пвт', 'о', 
    'фарм', 'ллс', 'аг', 'завод', 'кг', 'фармацевтическая', 'в', 
    'ве', 'б', 'д', 'м', 'н', 'к', 'ш', 'ас', 'гбуз', 'де', 'сп', 
    'др', 'х', 'мед'
])

# Функция для очистки текста с учетом лишних слов
def clean_text(text):
    if pd.isna(text):
        return ''
    text = text.lower().strip()
    text = re.sub(r'\s+', ' ', text)  # Замена множества пробелов на один пробел
    text = re.sub(r'[^а-яa-z0-9\s]', '', text)  # Удаление специальных символов, кроме русских и английских букв, цифр и пробелов
    tokens = re.findall(r'\b\w+\b', text)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(filtered_tokens).strip()
```

```python
# Функция для поиска наилучшего совпадения
def match_manufacturer(manufacturer, clients_df, threshold=85):
    matches = process.extract(manufacturer, clients_df['title (очищенный)'], limit=1)
    if matches and matches[0][1] >= threshold:
        matched_row = clients_df[clients_df['title (очищенный)'] == matches[0][0]]
        if not matched_row.empty:
            original_title = matched_row['title'].values[0]
            manufacturer_id = matched_row['id'].values[0]
            return original_title, matches[0][1], manufacturer_id
    return None, None, None

# Создание новых колонок для хранения результатов матчинга
unique_manufacturers_df_2['% матчинга'] = None
unique_manufacturers_df_2['match_manufacturer'] = None
unique_manufacturers_df_2['manufacturer_id'] = None
```

```python
import pandas as pd
from tqdm import tqdm
from transformers import BertModel, BertTokenizer
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Загрузка данных
unique_manufacturers_df = pd.read_csv('/mnt/d/Doc_To_Date/20240702/unique_matched_manufacturers_v_2.csv')
clients_df = pd.read_csv('/mnt/d/Doc_To_Date/20240703/pharmhub_public_Clients.csv')

# Инициализация модели и токенизатора BERT
model_name = 'bert-base-multilingual-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Определение стоп-слов
stop_words = set([
    'лтд', 'с', 'а', 'ооо', 'фарма', 'гмбх', 'ао', 'фармасьютикал', 'ко', 'п', 'л', 'и', 'зао', 'р', 'инк', 'оао', 
    'пвт', 'о', 'фарм', 'ллс', 'аг', 'завод', 'кг', 'фармацевтическая', 'в', 'ве', 'б', 'д', 'м', 'н', 'к', 'ш', 
    'ас', 'гбуз', 'де', 'сп', 'др', 'х', 'мед'
])

# Функция для очистки текста
def preprocess_text(text):
    text = text.lower()
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# Очистка данных
unique_manufacturers_df['cleaned'] = unique_manufacturers_df['Производитель готовой ЛФ'].apply(preprocess_text)
clients_df['cleaned'] = clients_df['title'].apply(preprocess_text)

# Функция для получения эмбеддингов предложений
def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# Получение эмбеддингов для всех предложений
unique_manufacturer_embeddings = np.array([get_sentence_embedding(sentence) for sentence in unique_manufacturers_df['cleaned']])
client_embeddings = np.array([get_sentence_embedding(sentence) for sentence in clients_df['cleaned']])

# Вычисление косинусного сходства
cosine_similarities = cosine_similarity(unique_manufacturer_embeddings, client_embeddings)

# Поиск наилучших совпадений
threshold = 0.75
unique_manufacturers_df['% матчинга'] = None
unique_manufacturers_df['match_manufacturer'] = None
unique_manufacturers_df['manufacturer_id'] = None

for index, row in tqdm(unique_manufacturers_df.iterrows(), total=unique_manufacturers_df.shape[0], desc="Matching manufacturers"):
    similarities = cosine_similarities[index]
    best_score_idx = similarities.argmax()
    best_score = similarities[best_score_idx]
    
    if best_score >= threshold:
        best_match = clients_df.iloc[best_score_idx]
        unique_manufacturers_df.at[index, '% матчинга'] = best_score
        unique_manufacturers_df.at[index, 'match_manufacturer'] = best_match['title']
        unique_manufacturers_df.at[index, 'manufacturer_id'] = best_match['id']
    else:
        unique_manufacturers_df.at[index, '% матчинга'] = None
        unique_manufacturers_df.at[index, 'match_manufacturer'] = None
        unique_manufacturers_df.at[index, 'manufacturer_id'] = None

# Сохранение результата
unique_manufacturers_df.to_csv('/mnt/d/Doc_To_Date/20240702/unique_matched_manufacturers_bert.csv', index=False)
print("Результат сохранен в файл 'unique_matched_manufacturers_bert.csv'")
```
