```python
import numpy as np

# Шаг 3: Разделение списка GTIN на части
def split_list(lst, n):
    return [lst[i:i + n] for i in range(0, len(lst), n)]

gtin_chunks = split_list(gtin_list, 1000)  # Разделим на части по 1000 элементов
print(f"Количество частей: {len(gtin_chunks)}")
```

```python
def fetch_manuf_skus(gtin_chunks):
    all_results = []
    conn = connect_to_database()
    cursor = conn.cursor()
    for chunk in gtin_chunks:
        query = 'SELECT "GTIN" FROM "ManufSKUs"'
        cursor.execute(query)
        all_results.extend(cursor.fetchall())
    cursor.close()
    conn.close()
    return all_results

manuf_skus = fetch_manuf_skus(gtin_chunks)
```

```python
# Преобразуем данные из базы данных в датафрейм
columns = ['GTIN']
manuf_skus_df = pd.DataFrame(manuf_skus, columns=columns)

# Убедимся, что GTIN хранятся как строки и разделим значения через пробел
manuf_skus_df['GTIN'] = manuf_skus_df['GTIN'].astype(str)
manuf_skus_df['GTIN'] = manuf_skus_df['GTIN'].str.split(' ')

# Преобразуем список GTIN из базы данных в множество для быстрого поиска
gtin_in_db = set(gtin for sublist in manuf_skus_df['GTIN'] for gtin in sublist)

# Определим GTIN, которые есть и которых нет в базе данных
gtin_not_in_db = set(gtin_list) - gtin_in_db

print(f"Количество GTIN в базе данных: {len(gtin_in_db)}")
print(f"Количество GTIN отсутствующих в базе данных: {len(gtin_not_in_db)}")
```

```python
# Загрузка файла pharmhub_public_Clients.csv
clients_df = pd.read_csv('/mnt/d/Doc_To_Date/20240703/pharmhub_public_Clients.csv')

# Просмотр первых нескольких строк
print(clients_df.head())
```

```python
import pandas as pd
from fuzzywuzzy import process
from tqdm import tqdm

# Функция для поиска наилучшего совпадения
def match_manufacturer(manufacturer, clients_df, threshold=80):
    matches = process.extract(manufacturer, clients_df['title'], limit=1)
    if matches and matches[0][1] >= threshold:
        return matches[0][0], matches[0][1], clients_df[clients_df['title'] == matches[0][0]]['id'].values[0]
    return None, None, None

# Создание новых колонок для хранения результатов матчинга
nan_sku_status_df.loc[:, '% матчинга'] = None
nan_sku_status_df.loc[:, 'match_manufacturer'] = None
nan_sku_status_df.loc[:, 'manufacturer_id'] = None

# Поиск совпадений и заполнение новых колонок с отслеживанием прогресса
for index, row in tqdm(nan_sku_status_df.iterrows(), total=nan_sku_status_df.shape[0], desc="Matching manufacturers"):
    manufacturer = row['Производитель готовой ЛФ']
    match_name, match_score, match_id = match_manufacturer(manufacturer, clients_df)
    if match_name:
        nan_sku_status_df.loc[index, '% матчинга'] = match_score
        nan_sku_status_df.loc[index, 'match_manufacturer'] = match_name
        nan_sku_status_df.loc[index, 'manufacturer_id'] = match_id

# Просмотр первых нескольких строк для проверки
nan_sku_status_df
```
