```python
import re

# Адаптируем функцию для проверки названий товаров
def check_broken_name(name):
    # Проверяем условие 1: не содержать цифр после обработки
    if not re.search(r'\d', name):
        return True
    # Проверяем условие 2: пусто или только символы
    if re.search(r'^\s*$', name):
        return True
    # Проверяем условие 3: нет букв, только цифры
    if re.search(r'^\d+$', name):
        return True
    return False


```

```python
# Применяем функцию к нужной колонке и создаем новую колонку 'broken_name'
df['broken_name'] = df['Наименование товара на этикетке'].apply(check_broken_name)

# Проверим результаты применения функции
df[['Наименование товара на этикетке', 'broken_name']]
```

```python
import psycopg2
from psycopg2.extras import RealDictCursor

def fetch_sku_data(eans, conn_params):
    """Функция для получения данных skuId и title по списку EAN из БД."""
    query = """
    SELECT "EAN", "id" AS "skuId", "title"
    FROM "ManufSKUs"
    WHERE "EAN" = ANY(%s);
    """
    with psycopg2.connect(**conn_params) as conn:
        with conn.cursor(cursor_factory=RealDictCursor) as cur:
            cur.execute(query, (eans,))
            return cur.fetchall()

def update_dataframe(df2, conn_params):
    """Обновление датафрейма df2 данными из БД и подсчет статистики."""
    eans_to_check = df2[df2['skuId'].isna()]['ean'].unique()  # Уникальные EAN для проверки
    batch_size = 1000  # Определите размер пакета для запросов
    total_found_in_db = 0
    total_updated = 0
    
    for i in range(0, len(eans_to_check), batch_size):
        # Преобразовываем numpy.int64 в строку и удаляем ведущие нули
        batch_eans = [str(ean).lstrip('0') for ean in eans_to_check[i:i+batch_size]]
        sku_data = fetch_sku_data(batch_eans, conn_params)
        total_found_in_db += len(sku_data)  # Подсчитываем сколько EAN найдено в БД
        
        # Обновляем df2 данными из БД
        for data in sku_data:
            # Удаляем ведущие нули из EAN при сравнении
            mask = (df2['ean'].apply(lambda x: str(x).lstrip('0')) == data['EAN']) & df2['skuId'].isna()
            # Подсчитываем сколько успешно обновлено
            total_updated += mask.sum()
            df2.loc[mask, 'skuId'] = data['skuId']
            df2.loc[mask, 'title'] = data['title']
    
    # Подсчитаем сколько осталось пустых skuId после всех обновлений
    remaining_empty_skuId = df2['skuId'].isna().sum()

    # Вывод статистики
    print(f"Всего найдено EAN в БД: {total_found_in_db}")
    print(f"Всего успешно обновлено строк: {total_updated}")
    print(f"Осталось строк с пустым skuId: {remaining_empty_skuId}")
    
    return df2

# Пример использования
df2_updated = update_dataframe(df2, conn_params)
print(df2_updated.head())
```

```python
import pandas as pd

# Загрузим файл, указывая разделитель и кодировку
df4 = pd.read_csv('/mnt/d/Doc_To_Date/20240513/updated_all_ls_for_match.csv', delimiter=';', encoding='utf-8')

# Проверяем наличие NaN в колонке 'ean'
if df4['ean'].isna().any():
    print("Обнаружены NaN в столбце 'ean'. Замена NaN на -1.")
    df4['ean'] = df4['ean'].fillna(-1)  # Заменяем NaN на -1

# Преобразование столбца 'ean' в int64
df4['ean'] = df4['ean'].astype('int64')

# Проверим, что данные успешно загружены, например, выведем первые несколько строк
df4.head()
```

```python
# Функция для преобразования ean в int64, заменяя NaN на -1
def convert_ean(df):
    if df['ean'].isna().any():
        print("Обнаружены NaN в столбце 'ean'. Замена NaN на -1.")
        df['ean'] = df['ean'].fillna(-1)  # Заменяем NaN на -1
    df['ean'] = df['ean'].astype('int64')
    return df
```

```python
# Применяем функцию к df3 и df4
df3 = convert_ean(df3)
df4 = convert_ean(df4)

# Слияние df4 и df3 для проверки и возможного обновления данных
df4_merged = df4.merge(df3, on='ean', suffixes=('', '_new'))

# Фильтрация и обновление строк по условиям
condition = (df4_merged['skuStatus'] == 'o') & (df4_merged['skuStatus_new'].isin(['g', 'y']))
updated_rows = df4_merged[condition]  # Строки, которые нужно обновить

# Подсчитаем количество обновленных строк
num_updated = updated_rows.shape[0]

# Обновляем df4 данными из df3, если условия выполнены
df5 = df4.copy()  # Создаем новый DataFrame для сохранения результатов
df5.update(updated_rows[list(df4.columns)])

# Вывод статистики обновлений
print(f"Количество обновленных строк: {num_updated}")

# Показываем начало нового DataFrame для проверки
print(df5.head())
```
